In our problem, we will be working with data regarding all citations! As such, we are removing those that exclude self-citations. We are also removing any strings in the dataset, as they do not help with the data analysis. 

```python
#restrict to "self-citations excluded" dataset
sci100k = sci100k_Data.drop(['authfull', 'inst_name', 'cntry', 'rank (ns)', 'nc9623 (ns)', 'h23 (ns)', 'hm23 (ns)', 'nps (ns)',
       'ncs (ns)', 'cpsf (ns)', 'ncsf (ns)', 'npsfl (ns)', 'ncsfl (ns)',
       'c (ns)', 'npciting (ns)', 'cprat (ns)', 'np6023 cited9623 (ns)', 'sm-subfield-1', 'sm-subfield-2','sm-subfield-1 count','rank sm-subfield-1','rank sm-subfield-1 (ns)'], axis = 1)
sci100k.head(20)
```

```python
##add in a feature actyr to reflect active years of research
sci100k['actyr'] = sci100k['lastyr'] - sci100k['firstyr']

##add in features npf (number of first authored papers) and ncf (total cites to first authored papers). 
#this is to create more features for us to analyse their correlation to rank
sci100k['npf'] = sci100k['cpsf'] - sci100k['nps']
sci100k['ncf'] = sci100k['ncsf'] - sci100k['ncs']

##we can also remove firstyr and lastyr since we have created actyr
sci100k = sci100k.drop(['firstyr', 'lastyr'], axis = 1)
```

```python
# # Creating a new dataframe and adding more interaction features
sci4 = pd.DataFrame(sci100k_impt['rank'])
sci4['hmp6023'] = sci100k_impt['np6023'] * sci100k_impt['hcomp23']
sci4['hhm23'] = sci100k_impt['hm23'] * sci100k_impt['hcomp23']
sci4['hnc9623'] = sci100k_impt['nc9623'] * sci100k_impt['hcomp23']
sci4['hncsf'] = sci100k_impt['ncsf'] * sci100k_impt['hcomp23']
sci4['hnpsfl'] = sci100k_impt['npsfl'] * sci100k_impt['hcomp23']
sci4['hncsfl'] = sci100k_impt['ncsfl'] * sci100k_impt['hcomp23']
sci4['hnpciting'] = sci100k_impt['npciting'] * sci100k_impt['hcomp23']
sci4['hnp6023_cited9623'] = sci100k_impt['np6023 cited9623'] * sci100k_impt['hcomp23']
```

```python

# Normalizing the data to further improve correlation
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
sci4_normalized = pd.DataFrame(scaler.fit_transform(sci4), columns=sci4.columns)


# Calculating the correlation after adding new features
plt.figure(figsize=(12, 8))
sb.heatmap(sci4_normalized.corr(method='spearman'), annot=True, cmap='Pastel1', linewidths=0.5)
plt.title('Correlation Heatmap of Normalized Interaction Features')
plt.show()
```

```python

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score

# Assuming sci4_log is your current DataFrame with features engineered and rank included

# 1. Data Preparation
X = sci4_log.drop(columns=['rank'])
y = sci4_log['rank']

# Standardizing features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Splitting data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# 2. Random Forest Regression
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.predict(X_test)

mse_rf = mean_squared_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)
print("Random Forest Regression - Mean Squared Error:", mse_rf)
print("Random Forest Regression - R-squared:", r2_rf)

# 3. Gradient Boosting Regression
gb_model = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
gb_model.fit(X_train, y_train)
y_pred_gb = gb_model.predict(X_test)

mse_gb = mean_squared_error(y_test, y_pred_gb)
r2_gb = r2_score(y_test, y_pred_gb)
print("\nGradient Boosting Regression - Mean Squared Error:", mse_gb)
print("Gradient Boosting Regression - R-squared:", r2_gb)

# 4. Support Vector Regression
svr_model = SVR(kernel='rbf', C=1.0, epsilon=0.1)
svr_model.fit(X_train, y_train)
y_pred_svr = svr_model.predict(X_test)

mse_svr = mean_squared_error(y_test, y_pred_svr)
r2_svr = r2_score(y_test, y_pred_svr)
print("\nSupport Vector Regression - Mean Squared Error:", mse_svr)
print("Support Vector Regression - R-squared:", r2_svr)

# 5. Feature Importance Visualization (Random Forest)
feature_importance_rf = pd.Series(rf_model.feature_importances_, index=X.columns)
important_features_rf = feature_importance_rf.nlargest(10)

print("\nTop 10 Important Features (Random Forest Regression):")
print(important_features_rf)

plt.figure(figsize=(12, 8))
sb.barplot(x=important_features_rf.values, y=important_features_rf.index)
plt.title("Top 10 Important Features from Random Forest")
plt.xlabel("Importance Score")
plt.ylabel("Features")
plt.show()

```

```python

X = sci100k_impt.drop(columns=['rank'])
y = sci100k_impt['rank']
```

```python
# 1. SelectKBest for Feature Selection
best_features = SelectKBest(score_func=mutual_info_regression, k=10)
X_best = best_features.fit_transform(X, y)
selected_features = X.columns[best_features.get_support()]
```

```python
# Creating a new DataFrame with the selected features
sci100k_selected = pd.DataFrame(X_best, columns=selected_features)
sci100k_selected['rank'] = y
```

```python
# 2. Polynomial Expansion of Selected Features
poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(sci100k_selected.drop(columns=['rank']))
poly_features = poly.get_feature_names_out(selected_features)
```

```python
sci100k_poly = pd.DataFrame(X_poly, columns=poly_features)
sci100k_poly['rank'] = y
```

```python
# 3. Interaction Terms
# Adding interaction terms selectively
sci100k_interactions = sci100k_selected.copy()
sci100k_interactions['interaction_h23_hm23'] = sci100k_selected['h23'] * sci100k_selected['hm23']
sci100k_interactions['interaction_np6023_nc9623'] = sci100k_selected['np6023'] * sci100k_selected['nc9623']
```

```python
# Combine all engineered features into one DataFrame
sci100k_combined = pd.concat([sci100k_poly, sci100k_interactions], axis=1)
```

```python
# 4. Regularization (Ridge Regression)
X_combined = sci100k_combined.drop(columns=['rank'])
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_combined)

ridge = Ridge(alpha=1.0)
ridge.fit(X_scaled, y)
```

```python
# Displaying feature coefficients to see which features are most important
feature_importance = pd.Series(ridge.coef_, index=X_combined.columns)
important_features = feature_importance.nlargest(10)
print("Top 10 Important Features:")
print(important_features)
```

```python
# Plotting correlation heatmap for top features
top_features = list(important_features.index) + ['rank']

plt.figure(figsize=(12, 8))
sb.heatmap(sci100k_combined[top_features].corr(method='spearman'), annot=True, cmap='Pastel1', linewidths=0.5)
plt.title("Correlation Heatmap of Top Combined Features Including Rank")
plt.show()
```
